```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F
```

```python
spark = SparkSession.builder.master('local').config('spark.driver.memory', '16g').appName(
    'Read_file').getOrCreate()
```

```python
top_5k = spark.read.parquet(
    '/Users/gabriele.sabato/Downloads/part-00000-5b77aa87-d7ed-47f7-9384-49b1bf9c5043-c000.snappy.parquet')
```

```python
top_5k.show()
```

```python
top_5k_del_weeks = (
    top_5k
        .withColumn('delivery_weeks', F.ceil(F.col('del_days') / 5.))
        .select('date',
                'item_code',
                'N_sales_tot',
                'item_Price',
                'GOI',
                'GOV',
                'NOI',
                'NOV',
                'voucher_value',
                'tot_costs',
                'Margin',
                'delivery_weeks')
)
```
